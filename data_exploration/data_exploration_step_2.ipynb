```python
from __future__ import absolute_import, division, print_function


import numpy
import os
import pandas
from pandas import DataFrame, Series
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams

wd = os.getcwd()
os.chdir('/Users/travisgreenwald/downloads/')
dfArrests = pandas.read_csv('JSS_RawArrests.csv')
dfArrests = DataFrame(dfArrests)
dfArrests.head()
```

## Top 10 days by number of arrests

```python
dateSlice = dfArrests['\tARREST_DATE\t']
dateSlice = pandas.to_datetime(dateSlice)
grpDate = dfArrests.groupby('\tARREST_DATE\t')
dailyCount = grpDate['ID\t'].count()
dailyCount = dailyCount.sort_values(ascending=False)
dailyCount = DataFrame(dailyCount)
dailyCount[:10]
```

## Line plot for dates vs number of arrests

```python
dailyCount = grpDate['ID\t'].count()
dailyCount = dailyCount['2015-1-1':]
dailyCount.plot(figsize = (20,6))
plt.xlabel('Date')
plt.ylabel('Number of arrests')
plt.title('Arrests by date')
```
